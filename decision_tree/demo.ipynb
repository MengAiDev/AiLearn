Decision Tree Task:
1. 基于iris数据集，建立决策树模型，评估模型
2. 可视化决策树结构
3. 修改min_sample_leaf，对比结果

```python
# Load the dataset
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
data = load_iris(as_frame=True)
df = data.frame

df.head()
```

```python
# define x and y
x = df.drop(columns='target', axis=1)
y = df['target']

print(x.shape)
print(y.shape)
```

```python
# Establish the Decision Tree Model

from sklearn import tree
dc_tree = tree.DecisionTreeClassifier(criterion='entropy',  min_samples_leaf=5)
dc_tree.fit(x, y)
```

```python
# Evaluate the model
y_predict = dc_tree.predict(x)
print(y_predict)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_predict)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Visualize the Decision Tree
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(12, 8))
plot_tree(dc_tree, filled=True, feature_names=x.columns, class_names=data.target_names)
plt.title("Decision Tree Visualization")
plt.show()
```

修改min_sample_leaf，对比结果

```python
# Change min_samples_leaf to 1
dc_tree_min1 = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=1)
dc_tree_min1.fit(x, y)

# Evaluate the new model
y_predict_min1 = dc_tree_min1.predict(x)
accuracy_min1 = accuracy_score(y, y_predict_min1)
print(f"Accuracy with min_samples_leaf=1: {accuracy_min1:.2f}")

# Visualize the new Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(dc_tree_min1, filled=True, feature_names=x.columns, class_names=data.target_names)
plt.title("Decision Tree Visualization with min_samples_leaf=1")
```
